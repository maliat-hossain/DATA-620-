### Hello Graph
#### For this assignment a graph has been created using networkx. Edges,nodes and weights have been attributed for the purpose of this graph. This graph explores employees in  a Data Analysis department and their connection with each other.

```python

import networkx as nx
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
import pandas as pd
```

```python
G = nx.Graph()
G.add_edge('Ana','Ben',weight=13,relation='Friend')
G.add_edge('Ben','Carol',weight=5,relation='Friend')
G.add_edge('Ben','Dan',weight=7,relation='Friend')
G.add_edge('Ema','Ben',weight=10,relation='Friend')
G.add_edge('Ema','Ana',weight=1,relation='Foe')
G.add_edge('Freya','Ben',weight=13,relation='Friend')
G.edges(data=True)
```

```python

G.add_node('Ana',role='Data_Scientist')
G.add_node('Ben',role='Data_Manager')
G.add_node('Carol',role='Data_Analyst_I')
G.add_node('Dan',role='Business_Analyst')
G.add_node('Ema',role='Data_Analyst_I')
G.add_node('Carol',role='Data_Analyst_II')

G.nodes(data=True)
```

```python
nx.draw_networkx(G, with_labels=True)
```
